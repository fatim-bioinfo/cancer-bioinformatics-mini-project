```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
data = {
    "Gene_A": [5.2, 8.1, 6.3, 9.5, 4.8, 7.2, 10.1, 3.9],
    "Gene_B": [2.1, 3.5, 2.8, 4.2, 1.9, 3.0, 4.5, 1.5],
    "Drug_Response": [70, 40, 60, 30, 80, 50, 20, 90],
    "status": ["Alive", "Dead", "Alive", "Dead", "Alive", "Dead", "Dead", "Alive"]
}
df = pd.DataFrame(data)
df
print(df)
mean_values = df.groupby("status")["Gene_A"].mean()
print(mean_values)

df.boxplot(column="Gene_A", by="status")
plt.suptitle("")
plt.show()

plt.scatter(df["Gene_A"], df["Drug_Response"])
plt.xlabel("Gene_A")
plt.ylabel("Drug_Response")
plt.show()
df["Gene_A"].corr(df["Drug_Response"])
```
